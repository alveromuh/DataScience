## Visualisasi Menggunakan PCA

```python
type(df[genes].values)
```

```python
pca = PCA(n_components=3)
pca_result = pca.fit_transform(df[genes].values)

df['pca-one'] = pca_result[:, 0]
df['pca-two'] = pca_result[:, 1]
df['pca-three'] = pca_result[:, 2]

print('Explained Variation per Principal Component: {}'.format(pca.explained_variance_ratio_))
```

```python
plt.figure(figsize=(22, 8))

pca_combs = [("pca-one", "pca-two"), ("pca-one", "pca-three"), ("pca-two", "pca-three")]

for i, comb in enumerate(pca_combs):
  ax = plt.subplot(1, 3, i + 1)
  sns.scatterplot(
      x=comb[0], y=comb[1],
      hue="characteristics_ch1.1.histopathological exam",
      palette=sns.color_palette("tab10", 4),
      data=df,
      legend="full",
      ax=ax
  )

  plt.legend(loc="upper right", framealpha=0.3)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    xs = df["pca-one"],
    ys = df["pca-two"],
    zs = df["pca-three"],
    c = df.groupby('characteristics_ch1.1.histopathological exam').ngroup()
)

ax.set_xlabel('pca-one')
ax.set_ylabel('pca-two')
ax.set_zlabel('pca-three')

plt.show()
```
