# Decision Tree: Kriteria Gini

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

# Inisialisasi list untuk menyimpan akurasi training dan test
training_accuracy = []
test_accuracy = []

def tree_scores(i):
    clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=i, random_state=42)

    clf_gini.fit(standardized_X_train, y_train)
    y_pred_train = clf_gini.predict(standardized_X_train)
    y_pred_test = clf_gini.predict(standardized_X_test)

    # Menghitung dan menyimpan akurasi training dan test
    training_accuracy.append(accuracy_score(y_train, y_pred_train))
    test_accuracy.append(accuracy_score(y_test, y_pred_test))

# Looping untuk mencetak akurasi pada setiap max depth
for i in range(1, 4):
    tree_scores(i)

import matplotlib.pyplot as plt

y = [1, 2, 3]

# plotting training accuracy
plt.plot(y, training_accuracy, label="training accuracy")

# plotting test accuracy
plt.plot(y, test_accuracy, label="test accuracy")

plt.xlabel('max depth')
plt.ylabel('accuracy')
plt.title('Training Accuracy and Test Accuracy')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)

clf_gini.fit(standardized_X_train, y_train)
y_pred_train = clf_gini.predict(standardized_X_train)
y_pred_test = clf_gini.predict(standardized_X_test)

# Menghitung akurasi training dan test
training_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

# Menampilkan nilai akurasi
print("Training Accuracy:", training_accuracy)
print("Test Accuracy:", test_accuracy)
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred_test, average='weighted')
print("F1-score:", f1)

overall_accuracy = accuracy_score(y_test, y_pred_test)
print("Overall Accuracy:", overall_accuracy)
```

```python
from sklearn.metrics import log_loss
from sklearn.tree import DecisionTreeClassifier


training_loss = []
test_loss = []

def tree_scores(i):
    clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=i, random_state = 42)


    clf_gini.fit(standardized_X_train, y_train)
    y_pred_gini = clf_gini.predict_proba(standardized_X_test)
    y_pred_train_gini = clf_gini.predict_proba(standardized_X_train)


    training_loss.append(log_loss(y_train, y_pred_train_gini))

    test_loss.append(log_loss(y_test, y_pred_gini))


for i in range(1,4):
    tree_scores(i)
```

```python
import matplotlib.pyplot as plt

y = [1, 2, 3]

# plotting training loss
plt.plot(y, training_loss, label = "training loss ")

# plotting test loss
plt.plot(y, test_loss, label = "test loss ")

plt.xlabel('max depth')

plt.ylabel('log-loss')
# Judul
plt.title('Training Loss and Test Loss')
plt.legend()

plt.show() # Menampilkan
```

# Feature Importance

Kriteria Gini

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

# X dalam bentuk array numpy 2D
X = np.array(X)

# y dalam bentuk array numpy 1D
z_new = np.array(z_new)

model = DecisionTreeClassifier(criterion='gini', max_depth=i, random_state=42)
model.fit(X, z_new)

# Menghitung baseline accuracy
baseline_accuracy = model.score(X, z_new)

# Menghitung feature importance menggunakan perturbation-based
feature_importance = np.zeros(X.shape[1])
for i in range(X.shape[1]):
    X_permuted = X.copy()
    X_permuted[:, i] = np.random.permutation(X_permuted[:, i])
    permuted_accuracy = model.score(X_permuted, z_new)
    feature_importance[i] = baseline_accuracy - permuted_accuracy

# Menormalisasi feature importance
feature_importance /= feature_importance.max()

# Mengurutkan fitur berdasarkan tingkat pentingnya
sorted_indices = np.argsort(feature_importance)[::-1]

# Nama-nama fitur
feature_names = ['LBE', 'LB', 'AC', 'FM', 'UC', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'DL',
                 'DS', 'DP', 'Width', 'Min', 'Max', 'Nmax', 'Nzeros', 'Mode', 'Mean',
                 'Median', 'Variance', 'Tendency']

# Menampilkan hasil
for idx in sorted_indices:
    print(f'{feature_names[idx]}: {feature_importance[idx]}')
```

```python
import matplotlib.pyplot as plt

# Menampilkan hasil dalam tabel plotting
plt.figure(figsize=(8, 6))
plt.bar(range(len(sorted_indices)), feature_importance[sorted_indices])
plt.xticks(range(len(sorted_indices)), [feature_names[i] for i in sorted_indices], rotation=90)
plt.xlabel('Fitur')
plt.ylabel('Tingkat Penting')
plt.title('Tingkat Penting Fitur Kriteria Gini')
plt.tight_layout()
plt.show()
```
